```python
import numpy as np
import matplotlib.pyplot as plt

weeks = np.arange(1, 9)
covid_cases = [500, 700, 1200, 2000, 3000, 4500, 6000, 8000] #1st subplot
vaccination_rate = [5, 10, 20, 35, 50, 65, 75, 85] #2nd subplot


fig,(sub1,sub2) = plt.subplots(2,1,figsize=(4,5))
sub1.plot(weeks, covid_cases, marker='o', linestyle='-',color="blue",label="covid-19 cases")
sub1.set_xlabel('Weeks')
sub1.set_ylabel('Covid cases (19)')
sub1.set_title('Weekwise Covid cases')
sub1.legend()



sub2.plot(weeks,vaccination_rate , marker='p', linestyle='-',color="green",label="Vaccination rate")
sub2.set_xlabel('Weeks')
sub2.set_ylabel('Vaccination Rate (%)')
sub2.set_title('Vaccination Progress')
sub2.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


days = np.arange(1, 11)
stock_price = [150, 152, 149, 155, 160, 162, 158, 165, 170, 175]  # In USD
trading_volume = [1.2, 1.4, 1.1, 1.8, 2.0, 2.5, 2.2, 2.7, 3.0, 3.5]  # In million shares


fig, sub1 = plt.subplots()
sub1.set_xlabel('Days')
sub1.set_ylabel('Stock Price(IN USD)', color='green')
sub1.plot(days, stock_price, marker='p', linestyle='-', color='blue', label='Stock Price')


sub2 = sub1.twinx() #create a twin axes sharing the x axis (now to sub1 it shares)
sub2.set_ylabel('Trading Volume (In Million Shares)', color='pink')
sub2.plot(days, trading_volume, marker='s', linestyle='--', color='yellow', label='Trading Volume')
plt.title('Stock Price and Trading Volume Over 10 Days')
plt.tight_layout()
plt.show()
```
