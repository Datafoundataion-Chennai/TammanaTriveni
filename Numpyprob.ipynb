```python
ÔøΩ
ÔøΩ Challenge 2: The Space Mission Resource Allocation üíº 
‚è≥ Time: 30 minutes 
‚è≥ Concepts Covered: Array creation, sum, max, standard deviation, transposition 
ÔøΩ
ÔøΩ Scenario: 
NASA is preparing a Mars mission ‚è≥ and needs to efficiently allocate oxygen, water, and 
food for astronauts over 6 months. Your job as a space engineer is to analyze and optimize 
resource consumption! 
ÔøΩ
ÔøΩ Task: 
 Create a (6, 3) NumPy array, where: ‚Äì ‚Äì 
Rows (0-5) represent months. 
Columns (0-2) represent oxygen, water, and food consumption (in tons). 
Compute total resources required for the entire mission. 
Identify which resource had the highest consumption in any given month and in 
total. 
Compute the standard deviation of each resource to measure consistency in 
consumption. 
Transpose the matrix to display the resource-wise breakdown by month.
```

```python
import numpy as np
rand=np.random.randint(1,20,(6,3))
print(rand)
```

```python
oxygen = np.sum(rand[:, 0])
water= np.sum(rand[:, 1])
food= np.sum(rand[:, 2])
print(f"Total resources needed(tons):Oxygen: {oxygen}, Water: {water}, Food: {food}")
print(f"Highest consumption in a month:")
max_values=np.max(rand,axis=0)
print(max_values)
max_months = np.argmax(rand, axis=0)+1 
print(max_months)
highest_resource_index = np.argmax([oxygen, water, food])
resource_names = ["Oxygen", "Water", "Food"]

print(f"Highest consumption in a month: {resource_names[np.argmax(max_values)]} ({max_values[np.argmax(max_values)]} tons in month {max_months[np.argmax(max_values)]})")
std_oxygen = round(np.std(rand[:, 0]), 1)
std_water = round(np.std(rand[:, 1]), 1)
std_food = round(np.std(rand[:, 2]),1)
print(f"Standard deviation of consumption: Oxygen: {std_oxygen}, Water: {std_water}, Food: {std_food}")
print("Transposed matrix:\n")
print(rand.T)
```

```python
import numpy as np
dataset = np.random.randint(10,50,(5,10))
print(dataset)
avg = (np.sum(dataset,axis =1))/10
avg = [round(a,1) for a in avg]
print("Average points per game:")
print(avg)
sums = np.sum(dataset,axis=1)
print()
print(f"Best-performing player: Player {np.where(sums==max(sums))[0][0]} (Total: {max(sums)} points)")
print(f"Worst-performing player: Player {np.where(sums==min(sums))[0][0]} (Total: {min(sums)} points)")
print()
print("Games with scores above 30")
for i in range(len(dataset)):
    mask = dataset[i]>30
    if len(dataset[i][mask]) != 0:
        print(f"Player {i+1}: Games {dataset[i][mask]}")
print()
temp = {}
for i in range(len(sums)):
    temp[i+1] = sums[i]
print(temp)
di = dict(sorted(temp.items(), key=lambda item: item[1],reverse = True))
i =1
print("Sorted Players by Total Points")
for j in di.keys():
    print(f"{i}. Player {j} - {di[j]} points")
    i+=1
```
